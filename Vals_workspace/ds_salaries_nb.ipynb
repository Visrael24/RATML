```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import tree
import statistics
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/ds_salaries.csv')
pd.set_option('display.max_columns', None)
```

```python
df.drop(['salary', 'salary_currency'], inplace=True, axis=1)
```

```python
string_cols1 = []
numeric_cols1 = []
for key, val in df.iloc[44].to_dict().items():
    if isinstance(val, str):
        string_cols1.append(key)
    else:
        numeric_cols1.append(key)
```

```python
numeric_from_df1 = df[numeric_cols1]
dummy_vars1 = pd.get_dummies(df[string_cols1])
```

```python
X = pd.concat([numeric_from_df1, dummy_vars1], axis=1)
```

```python
X
```

```python
y = X['salary_in_usd']
X.drop(['salary_in_usd'], inplace=True, axis=1)
```

```python
model = DecisionTreeRegressor()
```

```python
model.fit(X, y)
```

```python
Xtrain, Xvalue, ytrain, yvalue = train_test_split(X, y, test_size=.02, random_state=42)
```

```python
train = model.score(Xtrain, ytrain)
value = model.score(Xvalue, yvalue)
```
